## general functions 

```python
def modelRun(model_train, model_val, model_test, y_train, y_val, y_test):
    ##create model
    model, criterion, optimizer = createModel(Feedforward)
    #initialize parameters
    best_loss = np.inf
    best_model = 0
    #train model
    for epoch in range(EPOCHS):
        #train the model on the train set
        outputs = model(model_train.float())
        #wrangle output and labels and apply loss function
        train_loss = criterion(outputs.squeeze(), y_train.float().squeeze())
        #apply backward propagation
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()

        with torch.no_grad():
            val_size = model_val.shape[0]
            outputs = model(model_val.float())
            val_loss = criterion(outputs.squeeze(), torch.tensor(y_val.values).float().squeeze())
            if val_loss < best_loss:
                best_model = copy.deepcopy(model)

    #test model
    size = model_test.shape[0]
    best_model.eval()
    with torch.no_grad():
        outputs = best_model(model_test.float())
        test_loss = criterion(outputs.squeeze(), torch.tensor(y_test.values).float().squeeze())
        fpr, tpr, thresholds = metrics.roc_curve(y_test, outputs.detach().numpy())
        AUC = metrics.auc(fpr, tpr)
        AUPRC = metrics.average_precision_score(y_test, outputs.detach().numpy())
        return (test_loss).numpy()/size , AUC, AUPRC
```

```python
def modelRunTransfer(source, target):
    
    ##learning parameter
    lam = 0.3


    #load data
    model_train_source, model_val_source, model_test_source, y_train_source, y_val_source, y_test_source = source
    model_train_target, model_val_target, model_test_target, y_train_target, y_val_target, y_test_target = target
    
    ##create model
    model_target, criterion_target, optimizer_target = createModel(Feedforward)

    #initialize parameters
    size_target = model_train_target.shape[0]
    best_loss_target = np.inf
    best_model_target = 0

    #train model
    for epoch in range(EPOCHS):
        #train the model on the target set
        outputs_target = model_target(model_train_target.float())
        #wrangle output and labels and get loss function
        train_loss_target = criterion_target(outputs_target.squeeze(), y_train_target.float().squeeze())

        ###Source
        #train the model on the source set
        outputs_source = model_target(model_train_source.float())
        #wrangle output and labels and getloss function
        train_loss_source = criterion_target(outputs_source.squeeze(), y_train_source.float().squeeze())

        #combine losses
        train_loss_target = train_loss_target + lam * train_loss_source

        ##train target
        optimizer_target.zero_grad()
        train_loss_target.backward()
        optimizer_target.step()

        ##val source
        with torch.no_grad():
            outputs_source = model_target(model_val_source.float())
            val_loss_source = criterion_target(outputs_source.squeeze(), torch.tensor(y_val_source.values).float().squeeze())

        ##val target
        with torch.no_grad():
            outputs_target = model_target(model_val_target.float())
            val_loss_target = criterion_target(outputs_target.squeeze(), torch.tensor(y_val_target.values).float().squeeze())
            val_loss_target = val_loss_target + lam * val_loss_source
            if val_loss_target < best_loss_target:
                best_model_target = copy.deepcopy(model_target)

    #test model on target
    size = model_test_target.shape[0]
    best_model_target.eval()
    with torch.no_grad():
        outputs_target = best_model_target(model_test_target.float())
        test_loss = criterion_target(outputs_target.squeeze(), torch.tensor(y_test_target.values).float().squeeze())
        fpr, tpr, thresholds = metrics.roc_curve(y_test_target, outputs_target.detach().numpy())
        AUC = metrics.auc(fpr, tpr)
        AUPRC = metrics.average_precision_score(y_test_target, outputs_target.detach().numpy())
        return (test_loss).numpy()/size , AUC, AUPRC
```

```python
def FedAvg(model, model_1, model_2):
    # Update the model with the mean of the differences
    for name, param in model.named_parameters():
            model.state_dict()[name].copy_(
                    torch.mean(torch.stack([model_1.state_dict()[name], 
                                            model_2.state_dict()[name]]), dim=0))
        
    # Assign averaged model to both sites
    model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())
    return model, model_1, model_2

def modelRunFedSite(model, criterion, optimizer, model_train,  y_train):
        for epoch in range(1):
            outputs = model(model_train.float())
            #wrangle output and labels and get loss function
            train_loss = criterion(outputs.squeeze(), y_train.float().squeeze())
            optimizer.zero_grad()
            train_loss.backward()
            optimizer.step()
        return model, criterion, optimizer

def modelRunFed(data_1, data_2):
    #load data
    model_train_1, model_val_1, model_test_1, y_train_1, y_val_1, y_test_1 = data_1
    model_train_2, model_val_2, model_test_2, y_train_2, y_val_2, y_test_2 = data_2
    
    ##create models
    model, _, _ = createModel(Feedforward)
    model_1, criterion_1, optimizer_1 = createModel(Feedforward)
    model_2, criterion_2, optimizer_2 = createModel(Feedforward)
    model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())

    best_loss = np.inf
    best_model = model

    #train model
    for epoch in range(EPOCHS):
        model_1, criterion_1, optimizer_1 = modelRunFedSite(model_1, criterion_1, optimizer_1, model_train_1,  y_train_1)
        model_2, criterion_2, optimizer_2 = modelRunFedSite(model_2, criterion_2, optimizer_2, model_train_2,  y_train_2)
        model, model_1, model_2 = FedAvg(model, model_1, model_2)

        ##val 1
        with torch.no_grad():
            outputs = model_1(model_val_1.float())
            val_loss = criterion_1(outputs.squeeze(), torch.tensor(y_val_1.values).float().squeeze())
            if val_loss < best_loss:
                best_model = copy.deepcopy(model)
    
    #test model on 1
    size = model_test_1.shape[0]
    best_model.eval()
    with torch.no_grad():
        outputs_1 = best_model(model_test_1.float())
        test_loss = criterion_1(outputs_1.squeeze(), torch.tensor(y_test_1.values).float().squeeze())
        fpr, tpr, thresholds = metrics.roc_curve(y_test_1, outputs_1.detach().numpy())
        AUC = metrics.auc(fpr, tpr)
        AUPRC = metrics.average_precision_score(y_test_1, outputs_1.detach().numpy())
    return (test_loss).numpy() / size , AUC, AUPRC
```

```python
def getResults(name1, name2, c):
    ##load data
    X1, y1 = loadData(name1, 1)
    X2, y2 = loadData(name2, 2)
    
    ##dataset 1 by itself
    model_train, model_val, model_test, y_train, y_val, y_test = prepData(X1,y1)
    loss_s, AUC_s, AUPRC_s = modelRun(model_train, model_val, model_test, y_train, y_val, y_test)
    ##train together    
    model_train, model_val, model_test, y_train, y_val, y_test = prepDataJoint(X1,y1, X2,y2)
    loss_j, AUC_j, AUPRC_j = modelRun(model_train, model_val, model_test, y_train, y_val, y_test)
    
    ## transfer learning
    source = prepData(X2,y2)
    target = prepData(X1,y1)
    loss_t, AUC_t, AUPRC_t = modelRunTransfer(source, target)

    ##fed learning
    data_1 = prepData(X1,y1)
    data_2 = prepData(X2,y2)
    loss_f, AUC_f, AUPRC_f = modelRunFed(data_1, data_2)
        

        
    ##save results
    losses = [loss_s, loss_j, loss_t, loss_f]
    AUCs = [AUC_s, AUC_j, AUC_t, AUC_f]
    AUPRCs = [AUPRC_s,  AUPRC_j, AUPRC_t, AUPRC_f]
        
    losses_df = pd.DataFrame(losses, index=['single', 'joint','transfer', 'federated']).T
    losses_df['cost'] = c
    AUC_df = pd.DataFrame(AUCs, index=['single', 'joint','transfer','federated']).T
    AUC_df['cost'] = c
    AUPRC_df = pd.DataFrame(AUPRCs, index=['single', 'joint','transfer','federated']).T
    AUPRC_df['cost'] = c
    return losses_df, AUC_df, AUPRC_df
```

```python
def grapher(df, score, path, range, save = False):
    df_change = 100 * (df[['joint', 'transfer', 'federated']] - df[['single']].values) / df[['single']].values
    df_change['cost'] = df['cost']
    sns.lineplot( x= df_change['cost'], y =df_change['joint'], alpha = 0.8, marker = 'o', label = 'Joint')
    sns.lineplot( x= df_change['cost'], y =df_change['transfer'], alpha = 0.8, marker = 'o', label = 'Transfer')
    sns.lineplot( x= df_change['cost'], y =df_change['federated'], alpha = 0.8, marker = 'o', label = 'Federated')
    plt.axhline(y=0, color='black', linestyle = '--', alpha = 0.5, label = 'Baseline')
    plt.xlabel('OT_cost', fontsize = 14)
    plt.ylabel(f'% Change in {score}', fontsize = 14)
    plt.legend(fontsize = 14)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    if save == True:
        plt.savefig(f'{path}{score}{range}_change.pdf')
    plt.show()
```

```python
##FF model
class Feedforward(torch.nn.Module):
        def __init__(self, input_size, hidden_size):
                super(Feedforward, self).__init__()
                self.input_size = input_size
                self.hidden_size  = hidden_size
                self.fc = nn.Sequential(
                        nn.Linear(self.input_size, self.hidden_size[0]),
                        nn.ReLU(),
                        nn.Linear(self.hidden_size[0], self.hidden_size[1]),
                        nn.ReLU(),
                        nn.Dropout(0.5),
                        nn.Linear(self.hidden_size[1], self.hidden_size[2]),
                        nn.ReLU(),
                        nn.Linear(self.hidden_size[2], self.hidden_size[3]),
                        nn.ReLU(),
                        nn.Linear(self.hidden_size[3], 1)
                )


                self.sigmoid = torch.nn.Sigmoid()

        def forward(self, x):
                output = self.fc(x)
                output = self.sigmoid(output)
                return output
```

```python
def createModel(modelClass):
    model = modelClass(28, [16,24, 20, 24])
    criterion = nn.BCELoss(reduction = 'sum')
    optimizer = torch.optim.Adam(model.parameters(), lr = 0.001, weight_decay = 0.0005)
    return model, criterion, optimizer
```

```python
def prepData(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=1)
    X_train, X_val, y_train, y_val = train_test_split(X_train,y_train, test_size = 0.3, random_state=1)
    # scale data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    X_test = scaler.transform(X_test)

    model_train = torch.tensor(X_train)
    model_val = torch.tensor(X_val)
    model_test = torch.tensor(X_test)

    y_train = torch.tensor(y_train.values)
    return model_train, model_val, model_test, y_train, y_val, y_test

def prepDataJoint(X1, y1, X2, y2):
    X1_train, X_test, y1_train, y_test = train_test_split(X1,y1, test_size = 0.3, random_state=2)
    X2_train, _ , y2_train, _ = train_test_split(X2,y2, test_size = 0.3, random_state=2)

    ##combine 2 datasets 
    X_ = pd.concat([X1_train, X2_train], axis = 0)
    y_ = pd.concat([y1_train, y2_train], axis = 0) 
    ##create val dataset
    X_train, X_val, y_train, y_val = train_test_split(X_,y_, test_size = 0.3, random_state=2)

    ## scale dataset
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    X_test = scaler.transform(X_test)

    ##move to tensors
    model_train = torch.tensor(X_train)
    model_val = torch.tensor(X_val)
    model_test = torch.tensor(X_test)
    y_train = torch.tensor(y_train.values)
    return model_train, model_val, model_test, y_train, y_val, y_test
```

```python
def loadData(name, dataset):
    path = '/gpfs/commons/groups/gursoy_lab/aelhussein/pfl/otcost_fl/data/credit/'
    ##load data
    X = pd.read_csv(f'{path}{name}.csv', sep = ' ', names = [i for i in range(29)])
    ##merge
    X = X.sample(frac = 1)
    if dataset == 1:
        X = X.iloc[:110]
    else:
        X = X
        
    ##get X and label
    y = X.iloc[:,-1]
    X = X.iloc[:,:-1]
    return X,y
```

## Gradient divergence

```python
def wrangleDictPlot(dictionary):
    #wrangle dict to plot df
    df= pd.DataFrame.from_dict(dictionary, orient = 'index')
    df_plot = pd.DataFrame()
    for c in costs:
        df_c = df.loc[c].apply(pd.Series)
        df_c['label'] = c
        df_plot = pd.concat([df_plot, df_c])
    df_plot = pd.melt(df_plot, id_vars = 'label')
    return df_plot
```

```python
def loadData(name):
    path = '/gpfs/commons/groups/gursoy_lab/aelhussein/pfl/otcost_fl/data/credit/'
    ##load data
    X = pd.read_csv(f'{path}{name}.csv', sep = ' ', names = [i for i in range(29)])
    ##merge
    X = X.sample(frac = 1)
    ##get X and label
    y = X.iloc[:,-1]
    X = X.iloc[:,:-1]
    return X,y
```

```python
def runGradDiv(name1, name2):
    ##load data
    X1, y1 = loadData(name1)
    X2, y2 = loadData(name2)

    ##dataset 1 by itself
    data_1 = prepData(X1,y1)
    data_2= prepData(X2,y2)
        #load data
    model_train_1, model_val_1, model_test_1, y_train_1, y_val_1, y_test_1 = data_1
    model_train_2, model_val_2, model_test_2, y_train_2, y_val_2, y_test_2 = data_2

    ##create models
    model, _, _ = createModel(Feedforward)
    model_1, criterion_1, optimizer_1 = createModel(Feedforward)
    model_2, criterion_2, optimizer_2 = createModel(Feedforward)
    model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())


    #save diversity result
    diversity = {}
    cosines = {}
    #train model
    for epoch in range(300):
        grads = {}
        grads['model_1'] = {}
        grads['model_2'] = {}
        
        for _ in range(1):
            outputs = model_1(model_train_1.float())
            #wrangle output and labels and get loss function
            train_loss_1 = criterion_1(outputs.squeeze(), y_train_1.float().squeeze())
            optimizer_1.zero_grad()
            train_loss_1.backward()   
            for name, param in model_1.named_parameters():
                grads['model_1'][name] = param.grad.clone().detach()
            optimizer_1.step()
            
        
        for _ in range(1):
            outputs = model_2(model_train_2.float())
            #wrangle output and labels and get loss function
            train_loss_2 = criterion_2(outputs.squeeze(), y_train_2.float().squeeze())
            optimizer_2.zero_grad()
            train_loss_2.backward()    
            for name, param in model_2.named_parameters():
                grads['model_2'][name] = param.grad.clone().detach()
            optimizer_2.step()
            
        for name, param in model.named_parameters():
                model.state_dict()[name].copy_(
                        torch.mean(torch.stack([model_1.state_dict()[name], 
                                                model_2.state_dict()[name]]), dim=0))
            
        model_1_w = torch.cat([w.flatten() for w in grads['model_1'].values()])
        model_2_w = torch.cat([w.flatten() for w in grads['model_2'].values()])
        num = torch.norm(model_1_w)**2  + torch.norm(model_2_w)**2
        denom = torch.norm(model_1_w + model_2_w)**2
        gradient_diversity_metric = (num/ denom).numpy()
        cosines_ = cosine(model_1_w.numpy().astype(float), model_2_w.numpy().astype(float))
        model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())
        diversity[epoch], cosines[epoch] = gradient_diversity_metric.reshape(1)[0], cosines_
    return diversity, cosines
```

```python
##run model on datasets
costs = [0.1, 0.2, 0.3, 0.4]
diversity_ot = {}
cosines_ot = {}
for c in costs:
    diversity_ot[c] = {}
    cosines_ot[c] = {}
    for i in range(100):
        diversity_ot[c][i], cosines_ot[c][i] = runGradDiv(f'data_1_{c}',f'data_2_{c}')
```
