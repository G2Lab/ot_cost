```python
def modelRun(model_train, model_val, model_test, y_train, y_val, y_test, batch_size, indices):
    train = data_utils.TensorDataset(model_train, y_train)
    train_loader = data_utils.DataLoader(train, batch_size=batch_size, shuffle=True)
    
    ## wrangle
    model_val.unsqueeze_(1)
    model_test.unsqueeze_(1)
    
    ##create model
    model, criterion, optimizer = createModel(LeNet5, indices)
    #initialize parameters
    best_loss = np.inf
    best_model = 0
    #train model
    for epoch in range(EPOCHS):
        for data in train_loader:
            #train the model on the train set
            model_train, y_train = data
            model_train.unsqueeze_(1)
            outputs = model(model_train.float())
            #wrangle output and labels and apply loss function
            train_loss = criterion(outputs.squeeze(), y_train.long().squeeze())
            #apply backward propagation
            optimizer.zero_grad()
            train_loss.backward()
            optimizer.step()

        with torch.no_grad():
            val_size = model_val.shape[0]
            outputs = model(model_val.float())
            val_loss = criterion(outputs.squeeze(), y_val.long().squeeze())
            if val_loss < best_loss:
                best_model = copy.deepcopy(model)

    #test model
    size = model_test.shape[0]
    best_model.eval()
    with torch.no_grad():
        outputs = best_model(model_test.float())
        test_loss = criterion(outputs.squeeze(), y_test.long().squeeze())
        test_loss = test_loss.cpu()
        outputs =  outputs.cpu().detach()
        accuracy = metrics.accuracy_score(y_test.cpu(), outputs.cpu().detach().numpy().argmax(axis = 1))
    return (test_loss).cpu().numpy()/size , accuracy
```

```python
def modelRunTransfer(source, target, batch_size, indices):
    
    ##learning parameter
    lam = 0.3

    #load data
    model_train_source, model_val_source, model_test_source, y_train_source, y_val_source, y_test_source = source
    model_train_target, model_val_target, model_test_target, y_train_target, y_val_target, y_test_target = target

    
    train_source = data_utils.TensorDataset(model_train_source, y_train_source)
    train_loader_source = data_utils.DataLoader(train_source, batch_size=batch_size, shuffle=True)
    
    train_target = data_utils.TensorDataset(model_train_target, y_train_target)
    train_loader_target = data_utils.DataLoader(train_target, batch_size=batch_size, shuffle=True)
    
    
    ##create model
    model_target, criterion_target, optimizer_target = createModel(LeNet5, indices)
    
    ## move model to gpu
    model_val_source.unsqueeze_(1)
    model_val_target.unsqueeze_(1)
    model_test_target.unsqueeze_(1)
    
    #initialize parameters
    size_target = model_train_target.shape[0]
    best_loss_target = np.inf
    best_model_target = 0

    #train model
    for epoch in range(EPOCHS):
        for data_source, data_target in zip(train_loader_source, train_loader_target):
            model_train_source, y_train_source = data_source
            model_train_source.unsqueeze_(1)
            model_train_target, y_train_target = data_target
            model_train_target.unsqueeze_(1)

            #train the model on the target set
            outputs_target = model_target(model_train_target.float())
            #wrangle output and labels and get loss function
            train_loss_target = criterion_target(outputs_target.squeeze(), y_train_target.long().squeeze())

            ###Source
            #train the model on the source set
            outputs_source = model_target(model_train_source.float())
            #wrangle output and labels and getloss function
            train_loss_source = criterion_target(outputs_source.squeeze(), y_train_source.long().squeeze())

            #combine losses
            train_loss_target = train_loss_target + lam * train_loss_source

            ##train target
            optimizer_target.zero_grad()
            train_loss_target.backward()
            optimizer_target.step()

        ##val source
        with torch.no_grad():
            outputs_source = model_target(model_val_source.float())
            val_loss_source = criterion_target(outputs_source.squeeze(), y_val_source.long().squeeze())

        ##val target
        with torch.no_grad():
            outputs_target = model_target(model_val_target.float())
            val_loss_target = criterion_target(outputs_target.squeeze(), y_val_target.long().squeeze())
            val_loss_target = val_loss_target + lam * val_loss_source
            if val_loss_target < best_loss_target:
                best_model_target = copy.deepcopy(model_target)

    #test model on target
    size = model_test_target.shape[0]
    best_model_target.eval()
    with torch.no_grad():
 
        outputs_target = best_model_target(model_test_target.float())
        test_loss = criterion_target(outputs_target.squeeze(), y_test_target.long().squeeze())
        accuracy = metrics.accuracy_score(y_test_target.cpu(), outputs_target.cpu().detach().numpy().argmax(axis = 1))
    return (test_loss).cpu().numpy()/size , accuracy
```

```python
def FedAvg(model, model_1, model_2):
    # Compute the weighted average of the model parameters from both sites
    params_1 = model_1.state_dict()
    params_2 = model_2.state_dict()
    all_params = {}
    for name in params_1:
        if c < 0.1:
            w1, w2 = 0.2, 0.8
        else:
            w2, w1 = 0.2, 0.8
        all_params[name] = w1 * params_1[name] + w2 * params_2[name]
    
    # Update the model with the weighted average
    model.load_state_dict(all_params)
    
    # Assign averaged model to both sites
    model_1.load_state_dict(all_params)
    model_2.load_state_dict(all_params)
    
    return model, model_1, model_2

def modelRunFedSite(model, criterion, optimizer, train_loader):
        for epoch in range(1):
            for model_train, y_train in train_loader:
                model_train.unsqueeze_(1)
                outputs = model(model_train.float())
                #wrangle output and labels and get loss function
                train_loss = criterion(outputs.squeeze(), y_train.long().squeeze())
                train_loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        return model, criterion, optimizer

def modelRunFed(data_1, data_2, batch_size, indices):
    #load data
    model_train_1, model_val_1, model_test_1, y_train_1, y_val_1, y_test_1 = data_1
    model_train_2, model_val_2, model_test_2, y_train_2, y_val_2, y_test_2 = data_2


    train_1 = data_utils.TensorDataset(model_train_1, y_train_1)
    train_loader_1= data_utils.DataLoader(train_1, batch_size=batch_size, shuffle=True)
    
    train_2 = data_utils.TensorDataset(model_train_2, y_train_2)
    train_loader_2 = data_utils.DataLoader(train_2, batch_size=batch_size, shuffle=True)
    
    ##create models
    model, _, _ = createModel(LeNet5, indices[0])
    model_1, _, optimizer_1 = createModel(LeNet5, indices[0])
    model_2, _, optimizer_2 = createModel(LeNet5, indices[1])
    criterion_1 = nn.NLLLoss()
    criterion_2 = nn.NLLLoss()

    #initialise models
    model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())


    ##move
    model_val_1.unsqueeze_(1)
    model_test_1.unsqueeze_(1)    

    best_loss = np.inf
    best_model = model

    #train model
    for epoch in range(EPOCHS):
        model_1, criterion_1, optimizer_1 = modelRunFedSite(model_1, criterion_1, optimizer_1, train_loader_1)
        model_2, criterion_2, optimizer_2 = modelRunFedSite(model_2, criterion_2, optimizer_2, train_loader_2)
        model, model_1, model_2 = FedAvg(model, model_1, model_2)

        with torch.no_grad():
            outputs = model_1(model_val_1.float())
            val_loss = criterion_1(outputs.squeeze(), y_val_1.long().squeeze())
            if val_loss < best_loss:
                best_model = copy.deepcopy(model_1)

    
    #test model on 1
    size = model_test_1.shape[0]
    best_model.eval()
    with torch.no_grad():
        outputs_1 = best_model(model_test_1.float())
        test_loss = criterion_1(outputs_1.squeeze(), y_test_1.long().squeeze())
        accuracy = metrics.accuracy_score(y_test_1.cpu(), outputs_1.cpu().detach().numpy().argmax(axis = 1))
    return (test_loss).cpu().numpy() / size , accuracy
```

```python
def getResults(name1, name2, c, batch_size, indices):
    ##load data
    X1, y1 = loadData(name1, 1, indices[0])
    X2, y2 = loadData(name2, 2, indices[1])
    
    ##dataset 1 by itself
    model_train, model_val, model_test, y_train, y_val, y_test = prepData(X1,y1)
    loss_s, ACC_s = modelRun(model_train, model_val, model_test, y_train, y_val, y_test, batch_size, indices[0])
    ##train together    
    model_train, model_val, model_test, y_train, y_val, y_test = prepDataJoint(X1,y1, X2,y2)
    loss_j, ACC_j = modelRun(model_train, model_val, model_test, y_train, y_val, y_test, batch_size, indices[0])
    
    ## transfer learning
    source = prepData(X2,y2)
    target = prepData(X1,y1)
    loss_t, ACC_t, = modelRunTransfer(source, target,batch_size, indices[0])

    ##fed learning
    data_1 = prepData(X1,y1)
    data_2 = prepData(X2,y2)
    loss_f, ACC_f, = modelRunFed(data_1, data_2, batch_size, indices)
        

        
    ##save results
    losses = [loss_s, loss_j, loss_t, loss_f]
    ACCs = [ACC_s, ACC_j, ACC_t, ACC_f]
        
    losses_df = pd.DataFrame(losses, index=['single', 'joint','transfer', 'federated']).T
    losses_df['cost'] = c
    ACC_df = pd.DataFrame(ACCs, index=['single', 'joint','transfer','federated']).T
    ACC_df['cost'] = c
    return losses_df, ACC_df
```

```python
def grapher(df, score, path, range, save = False):
    df_change = 100 * (df[['joint', 'transfer', 'federated']] - df[['single']].values) / df[['single']].values
    df_change['cost'] = df['cost']
    sns.lineplot( x= df_change['cost'], y =df_change['joint'], alpha = 0.8, marker = 'o', label = 'Joint')
    sns.lineplot( x= df_change['cost'], y =df_change['transfer'], alpha = 0.8, marker = 'o', label = 'Transfer')
    sns.lineplot( x= df_change['cost'], y =df_change['federated'], alpha = 0.8, marker = 'o', label = 'Federated')
    plt.axhline(y=0, color='black', linestyle = '--', alpha = 0.5, label = 'Baseline')
    plt.xlabel('OT_cost', fontsize = 14)
    plt.ylabel(f'% Change in {score}', fontsize = 14)
    plt.legend(fontsize = 14)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    if save == True:
        plt.savefig(f'{path}{score}{range}_change.pdf')
    plt.show()
```

```python
class LeNet5(nn.Module):
    def __init__(self, num_classes):
        super(LeNet5, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, 6, kernel_size=5, stride=1, padding=0),
            nn.BatchNorm2d(6),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(6, 16, kernel_size=5, stride=1, padding=0),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2))
        self.fc = nn.Linear(256, 120)
        self.relu = nn.ReLU()
        self.fc1 = nn.Linear(120, 84)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(84, num_classes)
        self.softmax = torch.nn.Softmax(dim = 1)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        out = self.relu(out)
        out = self.fc1(out)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.softmax(out)
        return out
```

```python
def loadData(name, dataset, indices):
    path = '/gpfs/commons/groups/gursoy_lab/aelhussein/classes/pfl/otcost_fl/data/EMNIST/'
    ##load data
    data = np.load(f'{path}{name}.npz')
    X = data['data']
    y = data['labels']
    
    if dataset == 1:
        if len(indices) < 12:
            load_size = len(indices) * 10
        else:
            load_size = len(indices) * 20
    else:
        if len(indices) < 12:
            load_size = len(indices) * 40
        else:
            load_size = len(indices) * 80

    idx = np.random.choice(np.arange(X.shape[0]), load_size, replace=False)
    y = y[idx]
    X = X[idx]


    ## add test data (to be split later)
    Xtest, ytest = getTest(indices, size = 500)
    X = np.concatenate((X, Xtest), axis = 0)
    y = np.concatenate((y, ytest), axis = 0)

    ##replace values with 0, 1 etc.
    replace_dict = {i: x for (i,x) in zip(indices, range(len(indices)))}
    def replace_func(x):
        if x in replace_dict:
            return replace_dict[x]
        else:
            return x
    vectorized_replace = np.vectorize(replace_func)
    y = vectorized_replace(y)


    return X,y
```

```python
def getTest(indices, size = 500):

    ##load test
    images_full, labels_full = extract_test_samples('byclass')
    indices_x = getIndices(indices, labels_full, size)

    ##randomly permute and take fixed size of points
    np.random.seed(1)
    indices_x = np.random.permutation(indices_x)
    
    ind_1 = indices_x[:size]

    ##pull data and labels
    X = images_full[ind_1] / 255
    y = labels_full[ind_1]
    

    return X, y
```

```python
def getIndices(indices, labels_full, size):
    ##get number of indices
    num_ind = len(indices)
    ##get number per label
    num_per_label = int(size / num_ind) + 10
    indices_use = []
    ## loop through and pull indices
    for ind in indices:
        indices_use.extend(np.where(np.isin(labels_full, ind) == True)[0][:num_per_label])
    return indices_use
```

```python
def prepData(X,y):
    ##pull test set out
    X_test, y_test = X[-500:], y[-500:]
    ##create val
    X, y  = X[:-500], y[:-500]
    X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2, random_state=1)

    model_train = torch.tensor(X_train).to(device)
    model_val = torch.tensor(X_val).to(device)
    model_test = torch.tensor(X_test).to(device)

    y_train = torch.tensor(y_train).to(device)
    y_val = torch.tensor(y_val).to(device)
    y_test = torch.tensor(y_test).to(device)
    
    return model_train, model_val, model_test, y_train, y_val, y_test

def prepDataJoint(X1, y1, X2, y2):
    ##remove test set
    X_test, y_test = X1[-500:], y1[-500:]
    
    X1_train, y1_train = X1[:-500], y1[:-500]
    X2_train, y2_train = X2[:-500], y2[:-500]
    ##combine 2 datasets 
    X = np.concatenate([X1_train, X2_train], axis = 0)
    y = np.concatenate([y1_train, y2_train], axis = 0) 
    ##create val dataset
    X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2, random_state=1)
    
    ##move to tensors
    model_train = torch.tensor(X_train).to(device)
    model_val = torch.tensor(X_val).to(device)
    model_test = torch.tensor(X_test).to(device)

    y_train = torch.tensor(y_train).to(device)
    y_val = torch.tensor(y_val).to(device)
    y_test = torch.tensor(y_test).to(device)

    return model_train, model_val, model_test, y_train, y_val, y_test

            
```

```python
def createModel(modelClass, indices):
    model = modelClass(len(indices))
    model.to(device)
    criterion = nn.NLLLoss(reduction='sum')
    optimizer = torch.optim.Adam(model.parameters(), lr = 0.001, weight_decay = 0.0005)
    return model, criterion, optimizer
```

```python
def loadData(name, dataset, indices):
    path = '/gpfs/commons/groups/gursoy_lab/aelhussein/classes/pfl/otcost_fl/data/EMNIST/'
    ##load data
    data = np.load(f'{path}{name}.npz')
    X = data['data']
    y = data['labels']
    
    if len(indices) < 12:
        load_size = len(indices) * 40
    else:
        load_size = len(indices) * 80

    idx = np.random.choice(np.arange(X.shape[0]), load_size, replace=False)
    y = y[idx]
    X = X[idx]


    ## add test data (to be split later)
    Xtest, ytest = getTest(indices, size = 500)
    X = np.concatenate((X, Xtest), axis = 0)
    y = np.concatenate((y, ytest), axis = 0)

    ##replace values with 0, 1 etc.
    replace_dict = {i: x for (i,x) in zip(indices, range(len(indices)))}
    def replace_func(x):
        if x in replace_dict:
            return replace_dict[x]
        else:
            return x
    vectorized_replace = np.vectorize(replace_func)
    y = vectorized_replace(y)


    return X,y
```

```python
def runGradDiv(name1, name2, c, indices, batch_size):
    ##load data
    X1, y1 = loadData(name1, 1, indices[0])
    X2, y2 = loadData(name2, 2, indices[1])

    ##dataset 1 by itself
    data_1 = prepData(X1,y1)
    data_2= prepData(X2,y2)

    #load data
    model_train_1, model_val_1, model_test_1, y_train_1, y_val_1, y_test_1 = data_1
    model_train_2, model_val_2, model_test_2, y_train_2, y_val_2, y_test_2 = data_2
    
    ##create models
    model, _, _ = createModel(LeNet5, indices[0])
    model_1, _, optimizer_1 = createModel(LeNet5, indices[0])
    model_2, _, optimizer_2 = createModel(LeNet5, indices[1])
    criterion_1 = nn.NLLLoss()
    criterion_2 = nn.NLLLoss()
    model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())

    #load data
    train_1 = data_utils.TensorDataset(model_train_1, y_train_1)
    train_loader_1 = data_utils.DataLoader(train_1, batch_size=batch_size, shuffle=True)
    train_2 = data_utils.TensorDataset(model_train_2, y_train_2)
    train_loader_2 = data_utils.DataLoader(train_2, batch_size=batch_size, shuffle=True)


    #save diversity result
    diversity = {}
    cosines = {}
    #train model
    for epoch in range(100):
        grads = {}
        grads['model_1'] = {}
        grads['model_2'] = {}
        
        for _ in range(1):
            for  model_train_1, y_train_1 in train_loader_1:
                model_train_1.unsqueeze_(1)
                outputs = model_1(model_train_1.float())
                #wrangle output and labels and get loss function
                train_loss_1 = criterion_1(outputs.squeeze(), y_train_1.long().squeeze())
                train_loss_1.backward()   
                for name, param in model_1.named_parameters():
                    try:
                        grads['model_1'][name] += param.grad.clone().detach()
                    except:
                        grads['model_1'][name] = param.grad.clone().detach()

            optimizer_1.step()
            optimizer_1.zero_grad()
            
        
        for _ in range(1):
            for  model_train_2, y_train_2 in train_loader_2:
                model_train_2.unsqueeze_(1)
                outputs = model_2(model_train_2.float())
                #wrangle output and labels and get loss function
                train_loss_2 = criterion_2(outputs.squeeze(), y_train_2.long().squeeze())
                train_loss_2.backward()    
                for name, param in model_2.named_parameters():
                    try:
                        grads['model_2'][name] += param.grad.clone().detach()
                    except:
                        grads['model_2'][name] = param.grad.clone().detach()
            optimizer_2.step()
            optimizer_2.zero_grad()
            
        for name, param in model.named_parameters():
                model.state_dict()[name].copy_(
                        torch.mean(torch.stack([model_1.state_dict()[name], 
                                                model_2.state_dict()[name]]), dim=0))
            
        model_1_w = torch.cat([w.flatten() for w in grads['model_1'].values()])
        model_2_w = torch.cat([w.flatten() for w in grads['model_2'].values()])
        num = torch.norm(model_1_w)**2  + torch.norm(model_2_w)**2
        denom = torch.norm(model_1_w + model_2_w)**2
        gradient_diversity_metric = (num/ denom).numpy()
        cosines_ = cosine(model_1_w.numpy().astype(float), model_2_w.numpy().astype(float))
        model_1.load_state_dict(model.state_dict()), model_2.load_state_dict(model.state_dict())
        diversity[epoch], cosines[epoch] = gradient_diversity_metric.reshape(1)[0], cosines_
    return diversity, cosines
```
